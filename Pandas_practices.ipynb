```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Name': ['A','B','C','D','E','F','G','H','I','J'],
    'Age': [20, 22, 25, 27, 30, 19, 24, 28, 21, 26],
    'Marks': [45, 67, 80, 55, 90, 35, 60, 75, 50, 85],
    'City': ['BLR','HYD','BLR','DEL','HYD','DEL','BLR','DEL','HYD','BLR']
})
```

```python
df
```

```python
#1. Get students with Marks between 60 and 80

df[df['Marks'].between(60,80)]
```

```python
df[(df['Marks']>= 60) & (df['Marks']<=80)]
```

```python
#2. Get students from 'BLR' city with Marks > 70
df[(df['City'] == 'BLR') & (df['Marks']>70)]
```

```python
#3. Get Names of students whose Marks are either 45, 50, or 90
df[df['Marks'].isin([45,50,90])]
```

```python
#4. Get rows where Age is NOT between 20 and 25
df[~df['Age'].between(20,25)]
```

```python
#5Create a new column Result
#Marks ≥ 60 → "Pass"
#Marks < 60 → "Fail"
df['Result']= np.where(df['Marks']>=60,'pass','fail')
df
```

```python
#Using .loc[]
df.loc[df['Marks'] >= 60, 'Result'] = 'Pass'
df.loc[df['Marks'] < 60, 'Result'] = 'Fail'
```

```python
#6. Update Marks:
#If Marks < 50 → 0
#If Marks > 80 → 100
df.loc[df['Marks']<50,'Marks']=0
df.loc[df['Marks']>80,'Marks']=100
print(df)
```

```python
# Valid Dataset (one_to_many)
import pandas as pd

# Customers (ONE side)
customers = pd.DataFrame({
    'CustomerID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie']
})

# Orders (MANY side)
orders = pd.DataFrame({
    'CustomerID': [1, 1, 2, 3, 3],
    'OrderID': [101, 102, 103, 104, 105],
    'Amount': [250, 300, 150, 400, 500]
})

```

```python
customers
```

```python
orders
```

```python
merge = pd.merge(customers,orders,on='CustomerID',validate='one_to_many')
merge
```

```python
# Students many_to_many
students = pd.DataFrame({
    'StudentID': [1, 2, 3],
    'Name': ['A', 'B', 'C']
})

# Enrollments (bridge table)
enrollments = pd.DataFrame({
    'StudentID': [1, 1, 2, 3, 3],
    'CourseID': ['C1', 'C2', 'C1', 'C2', 'C3']
})

# Courses
courses = pd.DataFrame({
    'CourseID': ['C1', 'C2', 'C3'],
    'CourseName': ['Math', 'Science', 'English']
})

# Merge step-by-step
df = pd.merge(students, enrollments, on='StudentID', validate='one_to_many')
final = pd.merge(df, courses, on='CourseID', validate='many_to_one')

print(final)
```
